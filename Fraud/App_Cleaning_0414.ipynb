```python
#replace with record
#1.ssn = 999999999
#2.homephone = 9999999999
#3.address = 123 MAIN ST
#4. Dob 0229 to 0228
```

```python
data['dob'] = data['dob'].apply(str)
```

```python
for i in range(len(data['dob'])):
    if data['dob'].loc[i][-4:] == "0229":
        data['dob'].loc[i] = int(data['dob'].loc[i]) -1
```

```python
data["dob"].str.contains("r'(0229)$").sum()
```

```python
data['ssn'] = np.where(data['ssn'] == 999999999, data['record'], data['ssn'])
data['homephone'] = np.where(data['homephone'] == 9999999999, data['record'], data['homephone'])
data['address'] = np.where(data['address'] == "123 MAIN ST", data['record'], data['address'])
```
